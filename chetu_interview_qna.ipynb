```python
# write a logic to count of character form given string without using inbuild python fuctions like count
str1 = 'i am citizen of india'


def count_characters(string, char):
    count = 0
    for i in string:
        if i == char:
            count += 1
    return count

char_to_count = 'i'


result = count_characters(str1, char_to_count)

print(f'{char_to_count} = {result} times')
```
